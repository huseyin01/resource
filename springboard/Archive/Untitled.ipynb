```python
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(111)

t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2*np.pi*t)
line, = ax.plot(t, s, lw=2)

ax.annotate('local ppt', xy=(1, 1), xytext=(3, 1.5),
            arrowprops=dict(facecolor='green', shrink=0.05),)

ax.set_ylim(-2,2)
plt.show()
```
